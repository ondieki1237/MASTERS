```python
def sort(nums):
    # outer loop - each pass moves the next largest element to the end
    for i in range(len(nums) - 1, 0, -1):
        # inner loop - compare adjacent items
        for j in range(i):
            if nums[j] > nums[j + 1]:
                # swap
                temp = nums[j]
                nums[j] = nums[j + 1]
                nums[j + 1] = temp

# test the function
nums = [5, 3, 5, 9, 8, 1, 7, 3, 4, 6]
sort(nums)
print(nums)
```
